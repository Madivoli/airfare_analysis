DATA PROCESSING, CLEANING AND IMPUTATION

```python
import pandas as pd

airfare = pd.read_csv("C:\\Users\Madivoli Analytics\\OneDrive\\Documents\\Projects\\Airline Analysis\\airfare_report.csv", low_memory=False)

print(airfare.head())
```

```python
airfare.info()
```

```python
# Dropping unnecessary columns
columns_to_drop = ["tbl", "Geocoded_City1", "Geocoded_City2", "tbl1apk"]
airfare = airfare.drop(columns=columns_to_drop)

print("Columns after dropping:", airfare.columns.tolist())
```

```python
airfare.describe()
```

```python
# Validating numerical data

airfare.select_dtypes ("number")
```

```python
# Converting city columns to string/text

columns_to_convert = ["city1", "city2"]

# Convert all at once
for column in columns_to_convert:
    airfare[column] = airfare[column].astype(str)
    
print(airfare[columns_to_convert].dtypes)
```

```python

print(airfare.info())
```

```python
# Identify numeric columns 

numeric_columns = airfare.select_dtypes(include=['int64', 'float64']).columns


# Finding numeric columns with missing values

columns_with_missing_values = [col for col in numeric_columns if airfare[col].isna().any()]


# Imputing a summary statistics (mean)
for col in columns_with_missing_values:
    airfare[col] = airfare[col].fillna(airfare[col].mean())

print(airfare.isna().sum())
```

```python
# Columns to fill missing values for
columns_to_replace = ["carrier_lg", "carrier_low"]

# Replace missing values in EACH column with Unkonw
for col in columns_to_replace:
    airfare[col] = airfare[col].fillna('Unknown')

print(airfare.isna().sum())
```

```python
# Saving the cleaned DataFrame
airfare.to_csv(r"C:\Users\Madivoli Analytics\OneDrive\Documents\Projects\Airline Analysis\airfare_cleaned.csv", 
               index=False)


print(f"Data saved successfully! Shape: {airfare.shape}")
```
